```python
user_no=2
user_folder='../../../User2/'
last_index_even=10 # For even user numbers
last_index_odd=40  # For odd user numbers
increment=2        # Number for the increment of the user numbers
```

```python
SF= 128
```

```python
def get_images(user_folder,path):

    user_csv = os.listdir(user_folder)
    user_csv = user_csv[1:]
        
    for csv in user_csv:
        df = pd.read_csv(user_folder + csv)
        cols = df.columns
        cols = cols[:14]
        
        for col in cols:
            powerSpectrum, freqenciesFound, time, imageAxis = plt.specgram(df[col], Fs=SF)
            plt.axis('off')
            plt.savefig(path + "/" + col + csv[:-4] + "&" + '.jpg',bbox_inches='tight',pad_inches = 0)     
```

```python
def getfulldata(user_no,user_folder,last,increment):
    
    # Parameters :- User Number, Folder location you want to store the Data to, Last user number upto which you want to iterate,
    
    # Increment number: Number for the increment of the user numbers
    
    
    for i in range(user_no,last+1,increment):
        
        try:
            os.mkdir("../Dataset/" + "CNNImages/")
            print("Folder Created")
        except:
            print("Folder Exists")
            
            
        path= "../Dataset/" + "CNNImages/"
        
        get_images(user_folder[:-2] + str(i) +'/',path)
```
