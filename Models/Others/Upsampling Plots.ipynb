## Arousal Upsampled Plot:-

```python
import numpy as np
import matplotlib.pyplot as plt

N = 2
ind = np.arange(N)  
width = 0.27

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)


Zero = [248,507]
rects1 = ax.bar(ind, Zero, width, color='lightblue')
One = [507,507] 
rects2 = ax.bar(ind+width, One, width, color='g')

ax.set_ylabel('Number of Samples')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('0', '1') )
ax.set_xlabel('Labels')
ax.legend( (rects1[0], rects2[0]), ('Before Upsampling','After Upsampling') )

def autolabel(rects):
    for rect in rects:
        h = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1*h, '%d'%int(h),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
plt.title('Arousal Data Before vs After Upsampling', fontsize=20)
plt.show()
```
