```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.cm as cm
import sys
import extract_vector as ev
from mpl_toolkits.mplot3d import Axes3D
```

```python
liking_labels=ev.getLabelData('liking')[0:20].ravel()
dominance_labels=ev.getLabelData('dominance')[0:20].ravel()
valence_labels=ev.getLabelData('valence')[0:20].ravel()
arousal_labels=ev.getLabelData('arousal')[0:20].ravel()
```

```python
data1=liking_labels
data2=dominance_labels
n_bins = len(set(data1.flatten())) - 1

fig, ax = plt.subplots()
im = ax.hist2d(data1, data2,cmap='viridis',linewidths=0.1,color='white')
plt.xlabel("Liking")
plt.ylabel("Dominance")
plt.colorbar(im[3])
plt.show()
```

```python
data1=liking_labels
data2=valence_labels
n_bins = len(set(data1.flatten())) - 1

fig, ax = plt.subplots()
im = ax.hist2d(data1, data2,cmap='viridis',linewidths=0.1,color='white')
plt.xlabel("Liking")
plt.ylabel("Valence")
plt.colorbar(im[3])
plt.show()
```

```python
data1=liking_labels
data2=arousal_labels
n_bins = len(set(data1.flatten())) - 1

fig, ax = plt.subplots()
im = ax.hist2d(data1, data2,cmap='viridis',linewidths=0.1,color='white')
plt.xlabel("Liking")
plt.ylabel("Arousal")
plt.colorbar(im[3])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure(figsize=(12,7))
ax = fig.add_subplot(111, projection='3d')
x = liking_labels
y=  dominance_labels

hist, xedges, yedges = np.histogram2d(x, y, bins=10)


xpos, ypos = np.meshgrid(xedges[:-1] + 1.25, yedges[:-1] + 1.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

dx = dy = 0.5* np.ones_like(zpos)
dz = hist.ravel()

cmap = cm.get_cmap('viridis') 
max_height = np.max(dz)  
min_height = np.min(dz)
rgba = [cmap((k-min_height)/max_height) for k in dz] 

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average',color=rgba)

plt.title("Liking and Dominance vs Number of Videos")
ax.set_xlabel("Liking")
ax.set_ylabel("Dominance")
ax.set_zlabel('Number Of Videos')
plt.show()
```
